```python
import pandas as pd

df = pd.read_csv("data.csv", encoding='latin1',low_memory=False)
df.head()
```

```python
df.columns
```

```python
x = df['no2'].dropna().values
```

```python
import numpy as np

ar = 0.30
br = 1.5

z = x + ar * np.sin(br * x)
```

```python
m = np.mean(z)
sigma = np.var(z)

lam = 1 / (2 * sigma)
c = np.sqrt(lam / np.pi)

print("mean =", m)
print("lambda =", lam)
print("c =", c)
```
